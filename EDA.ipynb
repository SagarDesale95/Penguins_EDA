## Finding the column which are having missing values

```python
cols=[var for var in df.columns if df[var].isnull().mean()<0.05 and df[var].isnull]
cols
```

```python
df[cols].sample(5)
```

```python
len(df[cols].dropna())/len(df)
```

```python
new_df=df[cols].dropna()
df.shape,new_df.shape
```
